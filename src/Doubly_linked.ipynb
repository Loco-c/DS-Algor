```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None
        
```

```python
class Doubly_linked:
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1
    
    def print_list(self):
        temp = self.head
        while temp:
            print(temp.value)
            temp = temp.next
    
    def append(self,value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
        self.length += 1
        return True
    
    def pop(self):
        if self.length == 0:
            return None
        temp = self.tail
        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
            temp.prev = None
        self.length -= 1
        return temp
    
    def prepend(self, value):
        # create Node
        new_node = Node(value)
        # check if the list empty
        
        if self.length == 0:
            # head and tail equals new node
            self.head = new_node
            self.tail = new_node
        else:
            """
            The logic is below
            """
            # points at head node
            new_node.next = self.head
            # points at the new node
            self.head.prev = new_node
            """
            The new node is set to the new head node
            """
            self.head = new_node
        # increment by one 
        self.length += 1 
        return True
    
    # Removes the first value in list
    def pop_first(self):
        if self.length == 0:
            return None
        temp = self.head
        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None
            temp.next = None
        
        self.length -= 1
        return temp
    # Returns Index value
    def get(self, index):
        if index < 0 or index >= self.length:
            return None
        temp = self.head
        if index < self.length/2:
            for _ in range(index):
                temp = temp.next
        else:
            temp = self.tail
            for _ in range(self.length - 1, index, -1):
                temp = temp.prev
                
        return temp
        
    def set_value(self, index, value):
        # Obtains the index value of the number
        temp = self.get(index) 
        ''' Checks if temp is not empty
            and ensure that values not out of bound 
            of Doubly linked list 
        '''
        if temp is not None:
            temp.value = value
            return True
        return False
            
  
    def insert(self, index, value):
        # Edge Cases 
            if index < 0 or index >= self.length:
                return False
            if index == 0:
                return self.prepend(value)
            if index == self.length:
                return self.append(value)
            # Creates nodes
            new_node = Node(value)
            # checks index of node
            before = self.get(index - 1)
            # Creates node connection
            after = before.next
            # logic to connect nodes
            new_node.prev = before
            new_node.next = after
            before.next = new_node
            after.prev = new_node
            self.length += 1
            return True
    
    def remove(self, index):
        # Edge cases
        if index < 0 or index >= self.length:
            return None
        if index == 0:
            return self.pop_first()
        if index == self.length - 1:
            return self.pop()
        temp = self.get(index)
        # node connects
        temp.next.prev = temp.prev
        temp.prev.next = temp.next
        # removes previous value from list
        temp.next = None
        temp.prev = None
        self.length -= 1
        return temp                
        
```
